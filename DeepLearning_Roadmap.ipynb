## 1.3 Calculus Review: Derivatives and Gradients

Derivatives tell us how to update neural network weights during training.

```python
# Numerical derivative approximation
def numerical_derivative(f, x, h=1e-5):
    """Compute derivative using finite differences"""
    return (f(x + h) - f(x - h)) / (2 * h)

# Example functions
def f(x):
    return x**2

def g(x):
    return np.sin(x)

# Test derivatives
x = 2.0
print(f"f(x) = x^2")
print(f"  f({x}) = {f(x)}")
print(f"  f'({x}) ‚âà {numerical_derivative(f, x)}")
print(f"  Analytical f'(x) = 2x = {2*x}")

print(f"\ng(x) = sin(x)")
print(f"  g({x}) = {g(x)}")
print(f"  g'({x}) ‚âà {numerical_derivative(g, x)}")
print(f"  Analytical g'(x) = cos(x) = {np.cos(x)}")

# Visualize gradient descent
x_vals = np.linspace(-3, 3, 100)
y_vals = x_vals**2

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = x¬≤')
plt.scatter([x], [f(x)], color='red', s=100, zorder=5, label=f'Point at x={x}')

# Draw tangent line
slope = 2 * x  # derivative
tangent_y = f(x) + slope * (x_vals - x)
plt.plot(x_vals, tangent_y, '--', color='orange', label=f'Tangent (slope={slope})')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Derivative as the Slope of Tangent Line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 1.4 Chain Rule and Partial Derivatives

The chain rule is the foundation of backpropagation in neural networks.

```python
# Chain rule example: f(g(x))
def outer(u):
    return u**3

def inner(x):
    return 2*x + 1

def composed(x):
    return outer(inner(x))

# Compute derivative using chain rule
x = 2.0
u = inner(x)  # u = 2x + 1

# df/dx = df/du * du/dx
df_du = 3 * u**2  # derivative of u^3
du_dx = 2         # derivative of 2x + 1
df_dx = df_du * du_dx

print("Chain Rule Example: f(g(x)) where f(u) = u¬≥ and g(x) = 2x + 1")
print(f"At x = {x}:")
print(f"  u = g(x) = {u}")
print(f"  df/du = 3u¬≤ = {df_du}")
print(f"  du/dx = 2 = {du_dx}")
print(f"  df/dx = df/du * du/dx = {df_dx}")
print(f"  Numerical check: {numerical_derivative(composed, x)}")

# Partial derivatives for multi-variable functions
def f_multi(x, y):
    return x**2 + 3*x*y + y**2

def partial_x(f, x, y, h=1e-5):
    """Partial derivative with respect to x"""
    return (f(x + h, y) - f(x - h, y)) / (2 * h)

def partial_y(f, x, y, h=1e-5):
    """Partial derivative with respect to y"""
    return (f(x, y + h) - f(x, y - h)) / (2 * h)

x, y = 1.0, 2.0
print(f"\n" + "="*50)
print(f"Multi-variable function: f(x,y) = x¬≤ + 3xy + y¬≤")
print(f"At point ({x}, {y}):")
print(f"  ‚àÇf/‚àÇx = {partial_x(f_multi, x, y)}")
print(f"  Analytical: 2x + 3y = {2*x + 3*y}")
print(f"  ‚àÇf/‚àÇy = {partial_y(f_multi, x, y)}")
print(f"  Analytical: 3x + 2y = {3*x + 2*y}")

# Gradient (vector of all partial derivatives)
gradient = np.array([partial_x(f_multi, x, y), partial_y(f_multi, x, y)])
print(f"  Gradient ‚àáf = {gradient}")
```

## 1.5 Probability Basics

Understanding probability is essential for working with data distributions and model uncertainty.

```python
# Probability distributions
np.random.seed(42)

# Uniform distribution
uniform_samples = np.random.uniform(0, 1, 1000)

# Normal (Gaussian) distribution
normal_samples = np.random.normal(loc=0, scale=1, size=1000)

# Visualize distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(uniform_samples, bins=30, density=True, alpha=0.7, color='blue')
axes[0].set_title('Uniform Distribution U(0,1)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')
axes[0].grid(True, alpha=0.3)

axes[1].hist(normal_samples, bins=30, density=True, alpha=0.7, color='green')
axes[1].set_title('Normal Distribution N(0,1)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Expected value and variance
print("Uniform Distribution:")
print(f"  Mean (E[X]): {np.mean(uniform_samples):.4f} (theoretical: 0.5)")
print(f"  Variance: {np.var(uniform_samples):.4f} (theoretical: 0.0833)")

print("\nNormal Distribution:")
print(f"  Mean (E[X]): {np.mean(normal_samples):.4f} (theoretical: 0)")
print(f"  Variance: {np.var(normal_samples):.4f} (theoretical: 1)")
print(f"  Std Dev: {np.std(normal_samples):.4f} (theoretical: 1)")
```

## 1.6 Python Essentials for Deep Learning

Quick review of Python features commonly used in deep learning.

```python
# List comprehensions
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# Lambda functions
relu = lambda x: max(0, x)
print(f"ReLU(-5) = {relu(-5)}, ReLU(5) = {relu(5)}")

# Classes for neural network layers
class Layer:
    def __init__(self, input_size, output_size):
        self.weights = np.random.randn(input_size, output_size) * 0.01
        self.bias = np.zeros(output_size)
        
    def forward(self, x):
        return x @ self.weights + self.bias
    
    def __repr__(self):
        return f"Layer({self.weights.shape[0]} ‚Üí {self.weights.shape[1]})"

# Create a simple layer
layer = Layer(3, 2)
print(f"\nNeural Network Layer: {layer}")
print("Weights shape:", layer.weights.shape)
print("Bias shape:", layer.bias.shape)

# Test forward pass
input_data = np.array([1, 2, 3])
output = layer.forward(input_data)
print(f"Input: {input_data}")
print(f"Output: {output}")

# Dictionary for hyperparameters
config = {
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 10,
    'optimizer': 'adam'
}
print(f"\nModel Config: {config}")
```

```python
# Enhanced Progress Tracker with Detailed Metrics
from datetime import datetime
import matplotlib.pyplot as plt

class DetailedProgressTracker:
    def __init__(self):
        self.phases = {
            "Phase 1": {
                "name": "Mathematical Foundations",
                "status": "In Progress",
                "completion": 15,
                "start_date": "2026-01-23",
                "target_weeks": 4,
                "tasks_total": 8,
                "tasks_done": 1,
                "notes": "Started with linear algebra review"
            },
            "Phase 2": {
                "name": "PyTorch Proficiency",
                "status": "Not Started",
                "completion": 0,
                "target_weeks": 5,
                "tasks_total": 8,
                "tasks_done": 0,
                "notes": ""
            },
            "Phase 3": {
                "name": "From Scratch",
                "status": "Not Started",
                "completion": 0,
                "target_weeks": 4,
                "tasks_total": 8,
                "tasks_done": 0,
                "notes": ""
            },
            "Phase 4": {
                "name": "Vision Mastery",
                "status": "Not Started",
                "completion": 0,
                "target_weeks": 6,
                "tasks_total": 9,
                "tasks_done": 0,
                "notes": ""
            },
            "Phase 5": {
                "name": "NLP & LLMs",
                "status": "Not Started",
                "completion": 0,
                "target_weeks": 7,
                "tasks_total": 10,
                "tasks_done": 0,
                "notes": ""
            },
            "Phase 6": {
                "name": "Specialization",
                "status": "Not Started",
                "completion": 0,
                "target_weeks": 999,  # Ongoing
                "tasks_total": 8,
                "tasks_done": 0,
                "notes": ""
            }
        }
    
    def update(self, phase, completion, tasks_done, notes=""):
        """Update progress for a phase"""
        if completion == 100:
            status = "Complete"
        elif completion > 0:
            status = "In Progress"
        else:
            status = "Not Started"
        
        self.phases[phase]["status"] = status
        self.phases[phase]["completion"] = completion
        self.phases[phase]["tasks_done"] = tasks_done
        self.phases[phase]["notes"] = notes
    
    def show(self):
        """Display current progress"""
        print("=" * 80)
        print("üìö DEEP LEARNING MASTERY ROADMAP - PROGRESS DASHBOARD")
        print("=" * 80)
        print(f"Last Updated: {datetime.now().strftime('%Y-%m-%d %H:%M')}\n")
        
        total_completion = sum(p["completion"] for p in self.phases.values()) / len(self.phases)
        print(f"üéØ Overall Progress: {total_completion:.1f}%\n")
        
        for phase_num, (phase_key, data) in enumerate(self.phases.items(), 1):
            emoji = "‚úÖ" if data["status"] == "Complete" else "üöÄ" if data["status"] == "In Progress" else "‚è∏Ô∏è"
            print(f"{emoji} {phase_key}: {data['name']}")
            print(f"   Status: {data['status']} | Completion: {data['completion']}%")
            print(f"   Tasks: {data['tasks_done']}/{data['tasks_total']} | Target: {data.get('target_weeks', '?')} weeks")
            if data.get("notes"):
                print(f"   Note: {data['notes']}")
            print()
        
        print("=" * 80)
    
    def visualize(self):
        """Create a visual progress chart"""
        phases = list(self.phases.keys())
        completions = [self.phases[p]["completion"] for p in phases]
        colors = ['green' if c == 100 else 'orange' if c > 0 else 'lightgray' for c in completions]
        
        plt.figure(figsize=(12, 6))
        bars = plt.barh(phases, completions, color=colors, edgecolor='black')
        
        # Add completion percentage labels
        for i, (bar, comp) in enumerate(zip(bars, completions)):
            if comp > 0:
                plt.text(comp + 2, i, f'{comp}%', va='center', fontweight='bold')
        
        plt.xlabel('Completion (%)', fontsize=12)
        plt.title('Deep Learning Roadmap Progress', fontsize=14, fontweight='bold')
        plt.xlim(0, 110)
        plt.grid(axis='x', alpha=0.3)
        plt.tight_layout()
        plt.show()

# Initialize and display
progress = DetailedProgressTracker()
progress.show()
progress.visualize()
```
